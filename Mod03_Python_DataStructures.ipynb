8. What is a hash table, and how does it relate to dictionaries in Python?
    - A hash table is a data structure that stores key–value pairs and allows fast lookup, insertion, and deletion.
    - A hash function converts a key into an integer (hash value).
This hash value determines where the value is stored in memory. When you access a key, Python recomputes the hash and retrieves the value instantly.
    - Python dictionaries are implemented using hash tables, enabling fast access.
